Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook


x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]


output_notebook()


p = figure(title="Scatter Plot graph", x_axis_label='X-axis', y_axis_label='Y-axis')


p.circle(x, y, size=10, color="navy", alpha=0.8)


show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure ,output_file ,show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers['petal_length'] , flowers['petal_width'])
show(p)
```

```python
x = [2,3,4,5,6]
y = [3,4,5,6,7]
output_file("line.html")
p = figure(title = "line plot")
p.scatter(x,y,fill_color = "blue" , legend_label = 'blue points' , size = 20)
show(p)
```
